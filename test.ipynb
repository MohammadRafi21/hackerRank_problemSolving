```python
no_student = int(input())
students = []
for i in range(no_student):
    name = input()
    marks = float(input())
    newList = [name,marks]
    students.append(newList)


grade = []

for student in students:
    grade.append(student[1])

lowest = min(grade)
grade2 = []
for i in grade:
    if i != lowest:
        grade2.append(i)

second_lowest = min(grade2)
sec_student = []
for student in students:
    if student[1] == second_lowest:
        sec_student.append(student[0])
    else:
        continue
sec_student.sort()
for i in sec_student:
    print(i)
```
